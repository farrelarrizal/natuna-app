```python
import difflib

def compare_texts(text1, text2):
    """
    Compare two texts and return the differences.
    """
    text1_lines = text1.splitlines()
    text2_lines = text2.splitlines()

    differ = difflib.Differ()
    diff = list(differ.compare(text1_lines, text2_lines))
    
    return diff

def get_differences(diff):
    """
    Extract and return the specific differences from the diff.
    """
    differences = {'added': [], 'removed': [], 'changed': []}
    for line in diff:
        if line.startswith('+ '):
            differences['added'].append(line[2:])
        elif line.startswith('- '):
            differences['removed'].append(line[2:])
        elif line.startswith('? '):
            differences['changed'].append(line[2:])
    return differences

# Sample long texts
text_a = model

text_b = final_model

# Compare the texts
diff = compare_texts(text_a, text_b)
differences = get_differences(diff)

print("Added lines:")
for line in differences['added']:
    print(line)
    
print("\nRemoved lines:")
for line in differences['removed']:
    print(line)

print("\nChanged lines:")
for line in differences['changed']:
    print(line)
```
